```python
# connection à la base de donnees cassandra inca_test
import cassandra
from cassandra.cluster import Cluster

cluster=Cluster(['127.0.0.1'],port=9042)
session=cluster.connect()

session.execute('USE inca_test;')
rows = session.execute('select id,charac,scatcode,charac from inms_by_scat;')
print(rows)
print("")
for row in rows:
    print(row)
    print("")
```
